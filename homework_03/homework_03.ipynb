```python
# конвертер регистров
import re

def convert_case(text, force_format=None):
    """
    Конвертирует snake_case и PascalCase.

    Args:
        text (str): входная строка для преобразования.
        force_format (str, optional): принудительный формат вывода ('snake' или 'pascal').

    Returns:
        str: преобразованная строка.
    """
    if force_format:
        if force_format.lower() == 'snake':
            return to_snake_case(text)
        elif force_format.lower() == 'pascal':
            return to_pascal_case(text)
        else:
            raise ValueError("force_format должен быть 'snake' или 'pascal'")

    # Автоопределение формата
    if '_' in text and text.islower():
        return to_pascal_case(text)
    else:
        return to_snake_case(text)



def to_pascal_case(snake_text):
    """Преобразует snake_case в PascalCase."""
    words = snake_text.split('_')
    capitalized_words = [word.capitalize() for word in words]
    return ''.join(capitalized_words)



def to_snake_case(pascal_text):
    """Преобразует PascalCase в snake_case."""
    # Вставляем _ перед заглавными буквами, кроме первой
    snake_with_underscores = re.sub(r'(?<!^)(?=[A-Z])', '_', pascal_text)
    # Переводим в нижний регистр
    return snake_with_underscores.lower()
```

```python
# проверка даты на корректность
from datetime import datetime

def is_valid_date(date_str):
    """
    Проверяет валидность даты в формате DD.MM.YYYY.

    Args:
        date_str (str): Дата в формате "DD.MM.YYYY"

    Returns:
        bool: True, если дата существует, False — в противном случае
    """
    try:
        datetime.strptime(date_str, '%d.%m.%Y')
        return True
    except ValueError:
        return False

# Проверка примеров
print(is_valid_date('29.02.2000'))  # True
print(is_valid_date('29.02.2001'))  # False
print(is_valid_date('31.04.1962'))  # False
```

```python
# проверка на простое число
# сначала проверки на положительность и на саму 1
def is_simple(n):
    if n == 1:
        return True
    if n < 1:
        raise ValueError("Число должно быть положительным")
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True
```

```python
# программа для внесения данных пользователей
def is_valid_name(name):
    """Проверяет, что имя/фамилия состоят только из букв."""
    return name.isalpha()

def format_name(name):
    """Приводит имя/фамилию к формату: первая буква заглавная, остальные — строчные."""
    if not name:
        return name
    return name.capitalize()

def validate_age(age_str):
    """Проверяет, что возраст — число от 18 до 60."""
    try:
        age = int(age_str)
        if 18 <= age <= 60:
            return age
        else:
            print("Ошибка: возраст должен быть от 18 до 60 лет.")
            return None
    except ValueError:
        print("Ошибка: возраст должен быть целым числом.")
        return None

def format_id(user_id_str, existing_ids):
    """Форматирует ID: преобразует в целое число, дополняет до 8 знаков нулями, проверяет уникальность."""
    try:
        user_id = int(user_id_str)
        formatted_id = f"{user_id:08d}"
        if formatted_id in existing_ids:
            print("Ошибка: ID должен быть уникальным.")
            return None
        return formatted_id
    except ValueError:
        print("Ошибка: ID должен быть целым числом.")
        return None

def get_user_data(existing_ids):
    """Получает и валидирует данные одного пользователя. Возвращает кортеж (id, name, surname, age) или None при ошибке/отмене."""
    print("\nВведите данные пользователя (для завершения ввода оставьте любое поле пустым):")

    # Ввод и проверка имени
    name = input("Имя: ").strip()
    if not name:  # Пустое поле — завершение ввода
        return None, None, None, None
    if not is_valid_name(name):
        print("Ошибка: имя должно состоять только из букв.")
        return None, None, None, None
    name = format_name(name)

    # Ввод и проверка фамилии
    surname = input("Фамилия: ").strip()
    if not surname:  # Пустое поле — завершение ввода
        return None, None, None, None
    if not is_valid_name(surname):
        print("Ошибка: фамилия должна состоять только из букв.")
        return None, None, None, None
    surname = format_name(surname)

    # Ввод и проверка возраста
    age_str = input("Возраст: ").strip()
    if not age_str:  # Пустое поле — завершение ввода
        return None, None, None, None
    age = validate_age(age_str)
    if age is None:
        return None, None, None, None

    # Ввод и проверка ID
    user_id_str = input("ID: ").strip()
    if not user_id_str:  # Пустое поле — завершение ввода
        return None, None, None, None
    formatted_id = format_id(user_id_str, existing_ids)
    if formatted_id is None:
        return None, None, None, None

    return formatted_id, name, surname, age

def add_user(users_dict):
    """Добавляет пользователя в словарь после проверки данных. Возвращает True, если пользователь добавлен, False — если ввод завершён."""
    user_id, name, surname, age = get_user_data(users_dict.keys())

    # Если все значения None — пользователь хочет завершить ввод
    if user_id is None:
        return False

    # Добавляем пользователя в словарь
    users_dict[user_id] = (name, surname, age)
    print(f"Пользователь с ID {user_id} успешно добавлен.")
    return True

def print_users_table(users_dict):
    """Выводит словарь пользователей в виде таблицы."""
    if not users_dict:
        print("Нет данных для отображения.")
        return

    # Заголовки таблицы
    headers = ["ID", "Имя", "Фамилия", "Возраст"]

    # Определяем ширину колонок: начинаем с длины заголовков
    col_widths = [len(h) for h in headers]

    # Обновляем ширину колонок на основе данных
    for user_id, (name, surname, age) in users_dict.items():
        col_widths[0] = max(col_widths[0], len(user_id))
        col_widths[1] = max(col_widths[1], len(name))
        col_widths[2] = max(col_widths[2], len(surname))
        col_widths[3] = max(col_widths[3], len(str(age)))

    # Линия-разделитель
    separator = "+" + "+".join(["-" * (w + 2) for w in col_widths]) + "+"

    # Вывод таблицы
    print(separator)
    header_row = "| " + " | ".join(f"{h:^{w}}" for h, w in zip(headers, col_widths)) + " |"
    print(header_row)
    print(separator)

    # Выводим строки в отсортированном порядке по ID
    for user_id, (name, surname, age) in sorted(users_dict.items()):
        row_data = [user_id, name, surname, str(age)]
        row = "| " + " | ".join(f"{item:^{w}}" for item, w in zip(row_data, col_widths)) + " |"
        print(row)

    print(separator)

def main():
    """Основная функция программы."""
    users = {}
    print("Программа для ввода данных пользователей. Для завершения ввода оставьте любое поле пустым.")

    while True:
        if not add_user(users):
            break

    print("\nВвод завершён. Список пользователей:")
    print_users_table(users)

if __name__ == "__main__":
    main()
```
