## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
# про разворот строки, запрашиваем ввод данных
number = input('Введите число из пяти цифр: ') 

# Проверка на длину 
if len(number) != 5:  
print('Ошибка: введите ровно пять цифр') 
else:

num1 = number[0]
num234 = number[1:4]
num5 = number[4]

num432 = num234[::-1]

# Собираем
result = num1 + num432 + num5
print('Результат:', result)
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
# про выходных до отпуска, вводим дней всего осталось
a = int(input('Введите дней до отпуска  '))
if a%7==5:
  print(f'Выходных осталось:{a//7*2+1}')
elif a%7==6:
print(f'Выходных осталось:{((a-(a%7)+7)//7*2)}')
elif a%7==4 or a%7==3 or a%7==2 or a%7==1:
print(f'Выходных осталось:{((a-(a%7))//7*2)}')
elif a%7==0:
print(f'Выходных осталось:{a//7*2}')
else:
print(f'Выходных осталось:{a}')
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python

# задача про плитку шоколада
a = int(input('Укажите длину плитки: '))
b = int(input('Укажите ширину плитки '))
c = int(input('Можно мне вот столько кубиков отломить по прямой '))
if c < a*b and (c%a==0 or c%b==0):
print('Да, так можно!')
else:
print('Такой кусок отломить нельзя')
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
#  I - 1, V - 5, X - 10, L - 50, C - 100, D - 500, M - 1000
a = int(input('Я умею превращать арабские циферки в римские, не спрашивай, просто вводи: '))
m=a//1000%10
c=a//100%10
l=a//10%10
i=a%10

if (m!=None):
total0 = 'M' * m
if (c!=None):
if c==9:
total1='CM'
elif c==4:
total1='CD'
elif c>4:
total1='D'+'C'*(c-5)
else:
total1='C'*c
if (l!=None):
if l==9:
total2='XC'
elif l==4:
total2='XL'
elif l>4:
total2='L'+'X'*(l-5)
else:
total2='X'*l
if (i!=None):
if i==4:
total3='IV'
elif i>5 and i<9:
total3='V'+(i-5)*'I'
elif i==5:
total3='V'
elif i==9:
total3='IX'
else:
total3='I'*i
print(f'АБРА КАДАБРА {total0+total1+total2+total3}')
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
userNumb = input('Введите ваше число: ') 
charList = ['1','2','3','4','5','6','7','8','9','0','.']
# Проверка на положительное вещественное число 
isFu = False

#проходим каждый символ и сравниваем с листом
for i in userNumb:
if i not in charList:
break
else:
isFu = True

if isFu == True and userNumb.count('.') < 2:
print('Вещественное число') 
else:
print('НЕ!!! Вещественное число') 
```
